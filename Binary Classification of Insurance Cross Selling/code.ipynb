```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.base import BaseEstimator,TransformerMixin


class FeatureEncoder(BaseEstimator,TransformerMixin):
    
    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        encoder=OneHotEncoder()
        matrix = encoder.fit_transform(X[['Vehicle_Damage']]).toarray()

        column_name=["D","ND"]

        for i in range(len(matrix.T)):
            X[column_name[i]]=matrix.T[i]


        matrix=encoder.fit_transform(X[['Gender']]).toarray()

        column_name=["Female","Male"]

        for i in range(len(matrix.T)):
            X[column_name[i]]=matrix.T[i]

        matrix=encoder.fit_transform(X[['Vehicle_Age']]).toarray()

        column_name=["M12","l1","G2"]

        for i in range(len(matrix.T)):
            X[column_name[i]]=matrix.T[i]

        return X
```

```python
class Feature_drop(BaseEstimator,TransformerMixin):

    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        return X.drop(["Vehicle_Age","Gender","Vehicle_Damage"],axis=1,errors="ignore")
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([("featureencoder", FeatureEncoder()),
                    ("feature_drop",Feature_drop())])
```

```python
stratified_train_data = pipeline.fit_transform(stratified_train_data)
```
