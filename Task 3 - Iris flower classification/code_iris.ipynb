```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
train_data['species'] = label.fit_transform(train_data['species'])
```

```python
x = train_data.drop('species', axis=1)
y = train_data['species']
```

```python
train_data.head()
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import  train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=41)
```

```python
train_data.corr()
```

```python
plt.figure(figsize=(8, 6))  
sns.heatmap(train_data.corr(), cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
Logist = LogisticRegression(max_iter=12)
```

```python
Logist.fit(x_train,y_train)
```

```python
Logist.score(x_train, y_train)
```

```python
y_predict = Logist.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_predict)
accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(x_train,y_train)
```

```python
y_predict = knn.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_predict)
accuracy
```
