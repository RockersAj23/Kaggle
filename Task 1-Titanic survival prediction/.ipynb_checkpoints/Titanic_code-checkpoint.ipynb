```python
from sklearn.base import BaseEstimator,TransformerMixin
from sklearn.impute import SimpleImputer
class AgeImputer(BaseEstimator,TransformerMixin):

    def fit(self,X,y=None):
        return self

    def transform(self , X):
        imputer = SimpleImputer(strategy="mean")
        X['Age']= imputer.fit_transform(X[['Age']])
        return X
```

```python
from sklearn.preprocessing import OneHotEncoder

class FeatureEncoder(BaseEstimator,TransformerMixin):

    def fit(self , X , y=None):
        return self

    def transform(self,X):
        encoder= OneHotEncoder()
        matrix = encoder.fit_transform(X[['Embarked']]).toarray()

        column_name=["C","S","Q","N"]\

        for i in range(len(matrix.T)):
            X[column_name[i]]=matrix.T[i]

        matrix =encoder.fit_transform(X[['Sex']]).toarray()

        column_name=["Female","Male"]

        for i in range(len(matrix.T)):
            X[column_name[i]]=matrix.T[i]

        return X
```

```python
class Feature_drop(BaseEstimator,TransformerMixin):

    def fit(self,X,y=None):
        return self

    def transform(self,X):
        return X.drop(["Embarked","Name","Ticket","Cabin","Sex","N"],axis=1,errors="ignore")
        
```

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([("ageimputer",AgeImputer()),
                    ("featureencoder", FeatureEncoder()),
                    ("feature_drop",Feature_drop())])
```

```python
strat_train_set=pipeline.fit_transform(strat_train_set)
```

```python
strat_train_set
```

```python
strat_train_set.info()
```

```python
from sklearn.preprocessing import StandardScaler

X = strat_train_set.drop(["Survived"],axis=1)
Y = strat_train_set["Survived"]

scaler=  StandardScaler()

X_data = scaler.fit_transform(X)
Y_data = Y.to_numpy()
```

```python
X_data
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

clf = RandomForestClassifier()

Parameter_Grid = [
    {"n_estimators": [10,100,200,500],"max_depth":[None,5,10],"min_samples_split":[2,3,4]}
]

grid_search = GridSearchCV(clf , Parameter_Grid , cv=3 ,scoring="accuracy", return_train_score = True)
grid_search.fit(X_data , Y_data)
```

```python
final_clarifier = grid_search.best_estimator_
```

```python
final_clarifier
```
